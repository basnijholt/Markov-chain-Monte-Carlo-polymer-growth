# PERM Rosenbluth

###### Welcome to my Pruned-enriched Rosenbluth method for the simulations of polymers

```python
"""Import the packages that are used"""
from __future__ import division, print_function # adhere Python 3 syntax
import numpy as np
import matplotlib.pyplot as plt
import scipy.spatial
%matplotlib inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 14, 10  # default image size for this interactive session
```

Define all the parameters nesssesary to peform the MC simulation

```python
"""Define all the parameters nesssesary to peform the MC simulation"""
steps = 500     # the number of MC steps
N = 150          # number of atoms that is placed
N_theta = 9      # number of thetas that is chosen
N_max = 3*steps  # maximum number of polymers
epsilon = 0.25   # constant for the LJ-potential
sigma = 0.8; sigma6 = sigma**6; sigma12 = sigma**12 # constant for the LJ-potential

bifurcations = np.zeros(N)
prunings = np.zeros(N)
elsers = np.zeros(N)

cnt = np.zeros(N)
x_new = np.zeros((N_theta, 2))
x = np.zeros((N, 2))
end2end = np.zeros((N, N_max))
W_array = np.zeros((N, N_max))
x[1, 0] = 1.0  # the second atom at (1, 0)
```

```python
def mc(x, L, W, pe):
    """Perform an MC step.

    Parameters:
    -----------
    x : polymer
    L : length
    W : weight
    pe : Pruned-Enriched Rosenbluth Boolean True or False
    """
    # generate an equally spaced vector with intevals of 2pi/N_theta + a random off-set
    theta = 2 * np.pi * np.random.rand() + np.linspace(0, 2*np.pi, N_theta)

    # the positions of the new atom that is placed
    x_new[:, 0] = x[L, 0] + np.cos(theta)
    x_new[:, 1] = x[L, 1] + np.sin(theta)

    # calculate the weights of the new positions
    w = np.exp(-E(x, x_new, L))
    w_sum = np.sum(w)
    W *= w_sum/(0.75*N_theta)

    # choose a theta with a roulette wheel algoritm, this gives the index
    theta_idx = np.digitize(w_sum * np.random.rand(1), np.cumsum(w))
    if theta_idx == 9:
        theta_idx = np.random.randint(0, 8) # if all the weights are zero, chose a random theta

    # place the next atom at the theta that was found
    x[L+1, :] = x_new[theta_idx, :]

    # calculate the end to end distance
    end2end_abs = x[0, :] - x[L+1, :]
    end2end[L+1, cnt[L+1]] = np.sum(end2end_abs**2)

    # put the weight in an array
    W_array[L+1, cnt[L+1]] = W
    W_sum = np.sum(W_array[L+1, :cnt[L+1]])
    W_0 = np.sum(W_array[2, :cnt[2]]) # weight of the first few walkers

    # set the upper and lower limit
    up = 2.0*W_sum/W_0
    low = 1.2*W_sum/W_0
    W_new = W

    cnt[L+1] += 1
    if pe == False:
        if L < N-2:
            mc(x, L+1, W_new, pe)

    if pe == True:
        if cnt[L+1] > N_max-2 and L < N-2:
            mc(x, L+1, W_new, pe)
        elif L < N-2:
            if W > up: # Enrich
                bifurcations[L] += 1
                W_new = W*0.5
                mc(x, L+1, W_new, pe)   # clone 1
                W_new = W*0.5
                mc(x, L+1, W_new, pe)   # clone 2
            elif W < low: # Prune
                if np.random.rand(1) < 0.5: # kill it with 50% chance
                    W_new = 2*W
                    mc(x, L+1, W_new, pe)
                    prunings[L] += 1
            else:
                elsers[L] += 1
                mc(x, L+1, W_new, pe)

def E(x, x_new, L):
    """ The lennnard-Jones potential.

    Parameters:
    -----------
    x : polymer
    x_new : 9 new proposed positions for theta
    L : length
    """
    r = scipy.spatial.distance.cdist(x[0:L, :], x_new, 'euclidean')
    return np.sum(4*epsilon*(sigma12/(r**12)-sigma6/(r**6)), axis=0)
```

# Simulation

```python
W = 1
pe = False # Pruned-Enriched on or off
for i in range(steps):
    mc(x, 1, 1, pe)
```

# Data analysis

```python
end_to_end_avg = np.average(end2end[2:], weights=W_array[2:], axis=1)
```

Now I am going to plot the end to end distance a function of polymer length and fit an exponential curve to it of the form $a(N-1)^b$ in which $b$ should be equal to 1.5

```python
""" Fit an exponential curve ( a*(N-1)**0.75 ) to the data and find the coefficient a"""
from scipy.optimize import curve_fit
import sympy as sym

xx = range(2, N)
yy = end_to_end_avg

xx = np.array(xx) #transform your data in a numpy array, 
yy = np.array(yy) #so the curve_fit can work

def func(N, a, b):
    return a*(N-1)**b

popt, pcov = curve_fit(func, xx, yy)

print('a =', popt[0])


plt.subplot(311)
plt.plot(range(2,N), end_to_end_avg)
ax = plt.gca()
plt.title(r'Weighted square of the end-to-end length $R$ as a function of $N$', )
plt.xlabel(r'$N$')
plt.ylabel(r'$R^2$')
plt.xlim(2,N*1.3)
plt.ylim(end_to_end_avg[0],end_to_end_avg[N-3]*1.3)
ax.set_xscale('log')
ax.set_yscale('log')

plt.subplot(312)
plt.plot(range(2,N), np.average(end2end[2:], axis=1))
ax = plt.gca()
plt.title(r'Not weighted square of the end-to-end length $R$ as a function of $N$')
plt.xlabel(r'$N$')
plt.ylabel(r'$R^2$')
plt.xlim(2,N*1.3)
#plt.errorbar(range(2,N), end_to_end_avg, error)
plt.ylim(end_to_end_avg[0],end_to_end_avg[N-3]*1.3)
ax.set_xscale('log')
ax.set_yscale('log')


plt.subplot(313)
plt.plot(range(2,N), yy, 'ro',label="Original Data")
xs = sym.Symbol('N')    
tex = sym.latex(func(xs,*popt)).replace('$', '')
plt.title(r'$f(N)= %s$' %(tex), fontsize=16)
plt.xlabel('Length # of beads')
plt.plot(xx, func(xx, *popt), label="Fitted Curve") #same as line above
plt.legend(loc='upper left')
ax.set_xscale('log')
ax.set_yscale('log')
plt.show()


plt.show()
```
